```python
import graph_tool.all as gt
import os
import pickle as cp
import math
from graph_tool.topology import subgraph_isomorphism
```

```python

def orbit(m):
    ver=[v for v in m.vertices()]
    orbmem=[0 for v in m.vertices()]
    orb=[]
    index=0
    while ver:
        l=0
        orbt=set()
        for i in subgraph_isomorphism(m,m,generator=True):
            orbt.add(i[ver[0]])
            l+=1
        orb.append(orbt) 
        for i in orbt:
            orbmem[i]=index
            ver.remove(m.vertex(i))
        index+=1
    return orb,orbmem,l
def sbreak(m):
    sb_conditions={}
    vmaps=subgraph_isomorphism(m,m)
    while len(vmaps)>1:          
        orb=[]
        ver=[v for v in m.vertices()]
        while ver:
            orbt=set()
            for i in range(len(vmaps)):
                orbt.add(vmaps[i][ver[0]])
            orb.append(list(orbt))
            for j in orbt:
                ver.remove(m.vertex(j))
        olen=[len(i) for i in orb]
        imax=olen.index(max(olen))
        sb_conditions[orb[imax][0]]=orb[imax][1:]
        vmaps=[vm for vm in vmaps if vm[m.vertex(orb[imax][0])]==orb[imax][0]]
    return sb_conditions
```
